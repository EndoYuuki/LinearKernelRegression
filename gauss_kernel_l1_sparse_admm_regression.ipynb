```python
import numpy as np
import model as mdl
import matplotlib.pyplot as plt
```

```python
# dataset
n = 50
x = np.linspace(-3,3,n)
y_true = np.sin(np.pi*x)/(np.pi*x) + 0.1 * x
y = y_true + 0.2 * np.random.randn(n)
```

```python
# settings
theta_0 = np.ones(n)
z_0 = np.ones(n)
u_0 = np.ones(n)
print(z_0.shape, u_0.shape)
```

```python
model = mdl.L1GaussKernelADMMModel(theta_0, z_0, u_0, np.ones(n), 0.1)
model.train(x, y)
plt.plot(x,y, 'o', label="sample points")
plt.plot(x,y_true, label="true curve")
plt.plot(x,model(x), label="l1 gauss regression")
plt.legend()
plt.show()
```
